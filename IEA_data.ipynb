```python
import pandas as pd
import numpy as np
```

```python
#feed the data from the iea to create the dataframe
df2 = pd.read_excel("http://www.iea.org/statistics/IEA_HeadlineEnergyData.xlsx", sheetname = "TimeSeries_1971-2017", skiprows = 1, index = False)
df1 = pd.read_excel("https://www.rug.nl/ggdc/docs/pwt90.xlsx", sheetname = "Data")
```

```python
#get rid of information we don't need from df
df2 = df2.drop(['NoCountry', 'NoProduct', 'NoFlow'], axis=1)
df2 = df2[df2.Product == 'Total']
df2 = df2[(df2.Flow =='Total primary energy supply (ktoe)') | (df2.Flow =='Total final consumption (ktoe)') | (df2.Flow=='Total CO2 emissions - Fuel Combustion (Mt of CO2)')]
df2 = df2.sort_values(by = 'Country', ascending = True)
df2 = df2.reset_index()
df2 = df2.drop(['index'], axis = 1)
```

```python
Country1 = df1.country
Year1 = df1.year
GDP = df1.rgdpna
list_labels = ['Country', 'Year', 'GDP']
list_columns = [Country1, Year1, GDP]
zipped = list(zip(list_labels, list_columns))
data = dict(zipped)
dataframe1 = pd.DataFrame(data)
```

```python
df = df2.sort_values(by = 'Country', ascending = True)
Years2 = []
Countries2 = []
Total_primary_energy_supply = []
Total_final_consumption= []
Total_CO2_emissions_Fuel_Combustion = []

for index, row in df2.iterrows():
    if df.loc[index, 'Country'] not in Countries2:
        for y in range(1971, 2017):
            Countries2.append(str(df2.loc[index, 'Country']))
            Years2.append(y)
    if(df2.loc[index, 'Flow'] == 'Total primary energy supply (ktoe)'):
        for y in range(1971, 2017):
            Total_primary_energy_supply.append(df.loc[index, y])
    elif(df2.loc[index, 'Flow'] == 'Total final consumption (ktoe)'):
        for y in range(1971, 2017):
            Total_final_consumption.append(df.loc[index,y])
    elif (df2.loc[index, 'Flow']== 'Total CO2 emissions - Fuel Combustion (Mt of CO2)'):
        for y in range(1971, 2017):
            Total_CO2_emissions_Fuel_Combustion.append(df.loc[index, y])
```

```python
dataframe2 = pd.DataFrame({'Country': Countries2, 'Year': Years2, 'Total CO2 emissions': Total_CO2_emissions_Fuel_Combustion, 'Total primary energy supply (ktoe)': Total_primary_energy_supply, 'Total final consumption (ktoe)': Total_final_consumption}, columns=['Country', 'Year', 'Total CO2 emissions', 'Total primary energy supply (ktoe)', 'Total final consumption (ktoe)'])
```

```python
Cmplt_Countries = []
Cmplt_Years = []
Cmplt_GDP = []
Cmplt_TotalPrimaryEnergySupply = []
Cmplt_TotalFinalConsumption = []
Cmplt_TotalCO2EmissionsFuelCombustion = []
for indexa in range(0, 49):
    country = Country1[indexa]
    for x in range(0, len(dataframe1.Country)-1):
        #print(dataframe1.Country[x])
        if dataframe1.iloc[x, 0] == country:
            print(dataframe1.Country[x])
            print(country)
            indexc = x
            break
    for y in range(1950, 2017):
        Cmplt_Countries.append(country)
        Cmplt_Years.append(y)
        if y in Year1:
            Cmplt_GDP.append(dataframe1.iloc[indexc+(y-1950), 1])
        else:
            Cmplt_GDP.append('NaN')
        if (country in Countries2) & (y in dataframe2.Year):
            for x in range(0, len(dataframe2.Country)-1):
                if dataframe2.Country[x] == country:
                    index2c = x
                break
            index2y = index2c+(y-1971)
            Cmplt_TotalPrimaryEnergySupply.append(dataframe2.iloc[index2y, 2])
            Cmplt_TotalFinalConsumption.append(dataframe2.iloc[index2y, 3])
            Cmplt_TotalCO2EmissionsFuelCombustion.append(dataframe2.iloc[index2y, 4])
            Countries2.remove(country)
        else:
            Cmplt_TotalPrimaryEnergySupply.append('NaN')
            Cmplt_TotalFinalConsumption.append('NaN')
            Cmplt_TotalCO2EmissionsFuelCombustion.append('NaN')
    dataframe1=dataframe1[(dataframe1.Country != country)]
    dataframe2=dataframe2[(dataframe2.Country != country)]
```

```python
for item in range(0, len(Countries2)-1):
    #print(index)
    country = Countries2[item]
    for x in range(0, len(dataframe2.Country)-1):
        if dataframe2.iloc[x, 0] == country:
            indexc = x
            break
    for year in range(1950, 1970):
        Cmplt_Countries.append(country)
        Cmplt_Years.append(year)
        Cmplt_GDP.append('NaN')
        Cmplt_TotalPrimaryEnergySupply.append('NaN')
        Cmplt_TotalFinalConsumption.append('NaN')
        Cmplt_TotalCO2EmissionsFuelCombustion.append('NaN')
    for year in range(1971, 2017):
        indexy = indexc + (year - 1971)
        Cmplt_Countries.append(country)
        Cmplt_Years.append(year)
        Cmplt_GDP.append('NaN')
        Cmplt_TotalPrimaryEnergySupply.append(dataframe2.iloc[indexy, 2])
        Cmplt_TotalFinalConsumption.append(dataframe2.iloc[indexy, 3])
        Cmplt_TotalCO2EmissionsFuelCombustion.append(dataframe2.iloc[indexy, 4])
```

```python
list_labels_final = ['Country', 'Year', 'GDP', 'TPES', 'TFC', 'TCEFC']
list_columns_final=[Cmplt_Countries, Cmplt_Years, Cmplt_GDP, Cmplt_TotalPrimaryEnergySupply, Cmplt_TotalFinalConsumption, Cmplt_TotalCO2EmissionsFuelCombustion]
zipped_final = list(zip(list_labels_final, list_columns_final))
data_final = dict(zipped_final)
dataframe_final = pd.DataFrame(data_final)
print(dataframe_final)
```
